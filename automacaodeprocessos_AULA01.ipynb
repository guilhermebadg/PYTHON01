```python
# PASSO 4: calcular os indicadores (faturamento & qtde de produtos)  
import pandas as pd

tabela = pd.read_excel(r"C:\Users\HOME\Downloads\Vendas - Dez.xlsx")
display(tabela)
```

```python
quantidade = tabela["Quantidade"].sum()

faturamento = tabela["Valor Final"].sum()

print (quantidade)
print (faturamento)
```
